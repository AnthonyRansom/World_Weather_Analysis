```python
# Import the dependencies.
import pandas as pd
import numpy as np
from citipy import citipy
from config import weather_api_key
import requests
import os
import time
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(-90.000,90.000,size=2000)
lngs = np.random.uniform(-180.000,180.000,size=2000)
lat_lngs = zip(lats,lngs)
# Create a list of the Latitudes and Longitudes
coordinates = list(lat_lngs)
```

```python
# Declare a list for cities retrieved from citipy
cities = []
# Get the nearest city for each coordinates.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    # only add to cities list If the city is unique
    if city not in cities:
        cities.append(city)
# confrim numebr of cities in cities list
len(cities)
```

```python
# Declare variable to hold retrieved data
city_data = []

# Create Base URL and paramaters to be passed to API
weather_api_url = "https://api.openweathermap.org/data/2.5/weather"
params = {
    "units": "Imperial",
    "appid": weather_api_key
}

# Log start of data retrieval
print("-----------------------------")
print("Beginning Data Retrieval     ")
print("-----------------------------")

# create and reset counters
record_count = 1
set_count = 1
total_sets = int(np.ceil(len(cities)/50))
  
# Loop through all the cities in the cities list
for indx, city in enumerate(cities):
        
    # Group cities in sets of 50 for logging purposes
    if (indx % 50 == 0 and indx >= 50):
        set_count += 1
        record_count = 1
        #due to limitations of API free tier wait 61 seconds before processing next set
        print("Wait 60 seconds before next set")
        time.sleep(61)
        
    # add city query field to params
    params["q"] = city.replace(" ","+")
        
    # Print record currently being processed
    print(f"Processing Record {record_count}/50 of Set {set_count}/{total_sets} | {city}")
    # Add 1 to the record count
    record_count += 1
        
    # Run an API request and retrieve data
    # try block to catch exceptions and skip cities that cannot be found
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(weather_api_url,params=params).json()
    except:
        print("Problem with API request")
        
    # Parse out the needed data.
    try:
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity_perc = city_weather["main"]["humidity"]
        city_clouds_perc = city_weather["clouds"]["all"]
        city_wind_speed = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weather_descr = city_weather["weather"][0]["description"]
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity_perc,
                          "Cloudiness": city_clouds_perc,
                          "Wind Speed": city_wind_speed,
                          "Country": city_country,
                          "Current Description": city_weather_descr})

    # skip the city if API did not return any data or on any other error
    except:
        print("City not found. Skipping...")

# Log end of data retrieval
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the array of dictionaries city_data to Dataframe
city_data_df = pd.DataFrame(city_data)
new_column_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[new_column_order]
city_data_df.head()
```

```python
city_data_df.to_csv(os.path.join("WeatherPy_Database.csv"),index_label="City_ID")
```
